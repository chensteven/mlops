```python
import json
```

```python
filename = "dump/benjel_news/stories"
```

```python
with open(filename+".json", 'r') as f:
  data = json.load(f)
```

```python
list_of_submissions = data
```

```python
list_of_submissions
```

```python
for submission in list_of_submissions:
  print(submission['title'])
```

```python
import csv
with open(filename+".csv", "w+") as f:
    f = csv.writer(f)
    f.writerow(["title"])
    for x in list_of_submissions:
        print(x['title'])
        f.writerow([x["title"]])
```
